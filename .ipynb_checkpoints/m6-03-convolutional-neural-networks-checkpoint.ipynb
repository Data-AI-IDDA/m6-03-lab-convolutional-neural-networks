### SETUP

```python
# Import necessary libraries
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np

# Set device to GPU if available, else CPU
device = "cpu"
print(f"Using device: {device}")

# Set random seed for reproducibility
torch.manual_seed(42)
if device == "cuda":
    torch.cuda.manual_seed(42)
```

### TASK 1 - PART A

```python
# 1. Load a single CIFAR-10 image
transform_basic = transforms.ToTensor()
cifar10_train = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_basic)

# Get the first image and its label
img_tensor, label = cifar10_train[0]  # Shape: (3, 32, 32)
img_batch = img_tensor.unsqueeze(0)   # Add batch dimension -> Shape: (1, 3, 32, 32)

# 2. Create handcrafted filters
vertical_filter = torch.tensor([[-1., 0., 1.], 
                                [-1., 0., 1.], 
                                [-1., 0., 1.]])

horizontal_filter = vertical_filter.T  # Transpose of vertical

blur_filter = torch.ones((3, 3)) / 9.0 # 1/9 * ones

# Function to create a Conv2d layer with a specific handcrafted filter
def create_custom_conv(filter_2d):
    conv = nn.Conv2d(in_channels=3, out_channels=1, kernel_size=3, padding=1, bias=False)
    # Reshape filter to (1, 1, 3, 3) and repeat for 3 input channels
    # We divide by 3 to average the channel contributions
    weight = filter_2d.view(1, 1, 3, 3).repeat(1, 3, 1, 1) / 3.0
    conv.weight = nn.Parameter(weight)
    return conv

conv_vert = create_custom_conv(vertical_filter)
conv_horiz = create_custom_conv(horizontal_filter)
conv_blur = create_custom_conv(blur_filter)

# 3. Apply filters
with torch.no_grad():
    out_vert = conv_vert(img_batch).squeeze()  # Squeeze removes batch and channel dims -> (32, 32)
    out_horiz = conv_horiz(img_batch).squeeze()
    out_blur = conv_blur(img_batch).squeeze()

# Convert original image to NumPy for plotting (transpose to H, W, C)
img_np = img_tensor.permute(1, 2, 0).numpy()

# 4. Visualize the results
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

axes[0].imshow(img_np)
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(out_vert.numpy(), cmap="gray")
axes[1].set_title("Vertical Edge Detector")
axes[1].axis('off')

axes[2].imshow(out_horiz.numpy(), cmap="gray")
axes[2].set_title("Horizontal Edge Detector")
axes[2].axis('off')

axes[3].imshow(out_blur.numpy(), cmap="gray")
axes[3].set_title("Blur Filter")
axes[3].axis('off')

plt.tight_layout()
plt.show()
```

### Filter Highlights Description

Based on the manual application of the 3x3 filters to the CIFAR-10 image, here is what each filter highlights:

1. **Vertical Edge Detector:** This filter detects changes in pixel intensity along the x-axis. It highlights vertical lines and the left/right boundaries of objects in the image, while suppressing areas of uniform color.
2. **Horizontal Edge Detector:** This filter detects changes in pixel intensity along the y-axis. It emphasizes horizontal lines and the top/bottom edges of the objects. 
3. **Blur Filter:** This filter computes the average of the neighboring pixels, effectively smoothing the image. It reduces image noise and detail, creating a softer, slightly out-of-focus version of the original image.

### TASK 1 - PART B

```python
class TinyCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, padding=1)
        self.pool1 = nn.MaxPool2d(2)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        self.pool2 = nn.MaxPool2d(2)

tiny_model = TinyCNN()
dummy_x = torch.randn(8, 3, 32, 32)

print(f"Input shape: {dummy_x.shape}")
out_conv1 = tiny_model.conv1(dummy_x)
print(f"After conv1: {out_conv1.shape}")

out_pool1 = tiny_model.pool1(out_conv1)
print(f"After pool1: {out_pool1.shape}")

out_conv2 = tiny_model.conv2(out_pool1)
print(f"After conv2: {out_conv2.shape}")

out_pool2 = tiny_model.pool2(out_conv2)
print(f"After pool2: {out_pool2.shape}")
```

### TASK 2

```python
# Define basic transforms for Task 2
transform_t2 = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# Load Datasets and DataLoaders
train_dataset_t2 = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_t2)
val_dataset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_t2)

train_loader_t2 = DataLoader(train_dataset_t2, batch_size=128, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=128, shuffle=False, num_workers=2)

# Define the CNN Architecture
class CIFARClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        # Block 1
        self.block1 = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        # Block 2
        self.block2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        # Classifier
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 8 * 8, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 10)
        )

    def forward(self, x):
        x = self.block1(x)
        x = self.block2(x)
        x = self.classifier(x)
        return x

# Initialize model
model_t2 = CIFARClassifier().to(device)

# Compute parameter count
total_params = sum(p.numel() for p in model_t2.parameters() if p.requires_grad)
print(f"Total Trainable Parameters: {total_params:,}")
```

```python
def train_model(model, train_loader, val_loader, epochs=15):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)
    
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    
    for epoch in range(epochs):
        # Training Phase
        model.train()
        running_loss, correct, total = 0.0, 0, 0
        
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item() * inputs.size(0)
            _, preds = outputs.max(1)
            correct += preds.eq(labels).sum().item()
            total += labels.size(0)
            
        train_loss = running_loss / total
        train_acc = correct / total
        
        # Validation Phase
        model.eval()
        val_loss, val_correct, val_total = 0.0, 0, 0
        
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                
                val_loss += loss.item() * inputs.size(0)
                _, preds = outputs.max(1)
                val_correct += preds.eq(labels).sum().item()
                val_total += labels.size(0)
                
        val_loss = val_loss / val_total
        val_acc = val_correct / val_total
        
        # Save history
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        
        print(f"Epoch {epoch+1:02d}/{epochs} | "
              f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
              f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")
        
    return history

def plot_history(history, title="Training History"):
    epochs = range(1, len(history['train_loss']) + 1)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Loss Plot
    ax1.plot(epochs, history['train_loss'], label='Train Loss', marker='o')
    ax1.plot(epochs, history['val_loss'], label='Val Loss', marker='o')
    ax1.set_title(f"{title} - Loss")
    ax1.set_xlabel("Epochs")
    ax1.set_ylabel("Loss")
    ax1.legend()
    ax1.grid(True)
    
    # Accuracy Plot
    ax2.plot(epochs, history['train_acc'], label='Train Acc', marker='o')
    ax2.plot(epochs, history['val_acc'], label='Val Acc', marker='o')
    ax2.set_title(f"{title} - Accuracy")
    ax2.set_xlabel("Epochs")
    ax2.set_ylabel("Accuracy")
    ax2.legend()
    ax2.grid(True)
    
    plt.show()

# Run Training for Task 2
print("--- Starting Training for Task 2 (No Augmentation) ---")
history_t2 = train_model(model_t2, train_loader_t2, val_loader, epochs=15)
plot_history(history_t2, title="Task 2 Baseline")
```

### TASK 3

```python
# Define Augmented Transforms
train_tf_augmented = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.1, contrast=0.1),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
])

# Create new dataset and dataloader for augmented data
train_dataset_t3 = datasets.CIFAR10(root='./data', train=True, download=True, transform=train_tf_augmented)
train_loader_t3 = DataLoader(train_dataset_t3, batch_size=128, shuffle=True, num_workers=2)

# Initialize a NEW model instance with the exact same architecture
model_t3 = CIFARClassifier().to(device)

print("--- Starting Training for Task 3 (With Augmentation) ---")
history_t3 = train_model(model_t3, train_loader_t3, val_loader, epochs=15)
plot_history(history_t3, title="Task 3 With Data Augmentation")
```

```python
# Code to calculate and format the final comparison table
best_val_t2 = max(history_t2['val_acc'])
final_train_t2 = history_t2['train_acc'][-1]
final_val_t2 = history_t2['val_acc'][-1]
gap_t2 = final_train_t2 - final_val_t2

best_val_t3 = max(history_t3['val_acc'])
final_train_t3 = history_t3['train_acc'][-1]
final_val_t3 = history_t3['val_acc'][-1]
gap_t3 = final_train_t3 - final_val_t3

print("Comparison Summary:")
print(f"Task 2 (no augmentation) | Best Val Acc: {best_val_t2:.4f} | Train/Val Gap (Last Epoch): {gap_t2:.4f}")
print(f"Task 3 (with augmentation) | Best Val Acc: {best_val_t3:.4f} | Train/Val Gap (Last Epoch): {gap_t3:.4f}")
```

### Task 3: Data Augmentation Results & Interpretation

Based on the final comparison table and the generated training plots, here is what changed when applying data augmentation:

1. **Massive Reduction in Overfitting:** The most significant improvement is the stabilization of the model. In Task 2 (no augmentation), the Train/Val gap at the final epoch was **~8.9% (0.0892)**, indicating the model was starting to memorize the training data. In Task 3, this gap was virtually eliminated, shrinking to just **~0.76% (0.0076)**. The generated plots clearly show the Validation and Training curves tracking each other almost perfectly.
2. **Slower, Harder Learning (Regularization Effect):** You might notice that the "Best Val Acc" slightly dropped from **80.07% (Task 2)** to **78.09% (Task 3)**. This is a normal and expected behavior. Data augmentation makes the training images constantly change (cropping, flipping, color shifting), which makes the learning task much harder for the CNN. 

**Conclusion:** Data augmentation acted as an excellent regularizer. Because the learning task became harder, 15 epochs were not quite enough time for the augmented model to reach its absolute peak accuracy. However, by preventing overfitting (near-zero gap), we have built a much more robust and generalized model. If we were to train this model for 40-50 epochs instead of 15, Task 3 would easily surpass Task 2 in overall accuracy.
